Maintenant essayons de loop tout les musées de la page

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
path = "C:/Users/flori/Desktop/Master 1/Q2/Data Management"
driver = webdriver.Chrome(executable_path=os.path.join(path, 'chromedriver'))
url = "https://www.365.be/fr/recherche/?category=1062&range=150"
driver.get(url)
driver.implicitly_wait(10)
soup = BeautifulSoup(driver.page_source, 'html.parser')
attractions = soup.find_all("div", class_="card__body culturel")
name_attractions = []
for attraction in attractions:
    name = attraction.find("h3").text.strip()
    name_attractions.append(name)
print(name_attractions)

driver.quit()
```

avoir la ville des attractions 

```python
path = "C:/Users/flori/Desktop/Master 1/Q2/Data Management"
driver = webdriver.Chrome(executable_path=os.path.join(path, 'chromedriver'))
url = "https://www.365.be/fr/recherche/?category=1062&range=150"
driver.get(url)
driver.implicitly_wait(10)
soup = BeautifulSoup(driver.page_source, 'html.parser')
villes = soup.find_all("div", class_="card__body culturel")
name_villes = []
for ville in villes:
    name = ville.find("div").text.strip()
    name_villes.append(name)
print(name_villes)

driver.quit()
```

Création dataframe avec attraction et ville associés

```python
attraction_infos = []
for attraction in attractions:
    event = attraction.find("h3").text.strip()
    city = attraction.find("div").text.strip()
    
    attraction_info = {'Attraction' : event, 'City' : city}
    attraction_infos.append(attraction_info)
df = pd.DataFrame(attraction_infos)
df
```
