# Webscraping

We will here try to do the webscrapping on the website https://www.365.be/fr/. This website have all the different activities present in Wallonia and in the region of Brussels capital. We will first import the different package needed for this webscrapping.

```python
from IPython.display import display
import json
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = 700
from pprint import pprint
import re
import time
import os
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
pip install selenium
```

After that, we will collect the different informations given by the website. We will make a list that stores the information about attractions extracted from the website. Each code is for a different type of attractions. Each attraction's information is stored as a dictionary containing the name of the attraction, the type of attraction, the city where the attraction is and the price for each groups of persons. We will use selenium because we have here a dynamic website.

First Loop in order to have the different informations for Abbeys and religious monuments.

```python
path = "C:Users/flori/Desktop/Master 1/Q2/Data Management"
driver = webdriver.Chrome(executable_path=os.path.join(path, 'chromedriver'))
attractions_infos1 = []
url = "https://www.365.be/fr/recherche/?category=1062&range=150&subCategory=1101"
driver.get(url)
time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')

events = soup.find('div', class_='main-content__body').find('div', class_='row cards').find_all('article', class_='col-12 col-md-6 col-lg-6 col-xl-4')

for event in events:
    full_event_url = f"https://www.365.be{event.find('a')['href']}"
    driver.get(full_event_url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    try:
        attraction_name = soup.find('h1', class_="attraction-name").text.strip()
    except:
        attraction_name = "Not available"
    try:
        attraction_type = "Abbeys and religious monuments"
    except:
        attraction_type = "Not available"
    try:
        city_name = soup.find('span', {'itemprop': 'addressLocality'}).text.strip()
    except:
        city_name = "Not available"
    try:
        price_label = soup.find("span", class_="label", text="Prix").find_next_sibling("div").text.strip()
    except:
        price_label = "Not available"
    attractions_infos1.append({'Name' : attraction_name,'Type': attraction_type, 'City' : city_name, 'Price' : price_label})
print(attractions_infos1)
driver.quit()
```

In the following code, we are going to extract from the price column the price for adults and children. We will also detail for each price what the code must take into account in order to have the best representation of the prices for the two groups.

```python
df["Price"] = df["Price"].astype(str)

# Define the function to extract price from the label
def extract_price(category, price_label):
    if category == "adultes":
        pattern = f"(adulte|adultes|adult).+?(\d+(?:[\.,]\d{{1,2}})?)\s*(?:€|\(.*?€\))"
    elif category == "enfants":
        pattern = f"(enfant|enfants).+?(\d+(?:[\.,]\d{{1,2}})?)\s*(?:€|\(.*?€\))"
    else:
        pattern = None
        
    if pattern:
        match = re.search(pattern, price_label, flags=re.IGNORECASE)
        if match:
            return match.group(2).replace(',', '.')
    
    if "gratuit" in price_label.lower() or "entrée gratuite" in price_label.lower() or "entrée libre" in price_label.lower():
        return 0
    pp_pattern = r"\b(\d+(?:[,.]\d{1,2})?)\s*€?\s*pp\b"
    pp_match = re.search(pp_pattern, price_label, flags=re.IGNORECASE)
    if pp_match:
        return pp_match.group(1).replace(',', '.')
    if len(price_label.split()) == 1:
        return price_label.replace(',','.')    
    
    return None

# Extract prices for different categories
df["Adults"] = df["Price"].apply(lambda x: extract_price("adultes", x))
df["Children"] = df["Price"].apply(lambda x: extract_price("enfants", x))
```

The following code will remove the euro symbol that could be present with the values in the columns 'Adults' and 'Children', it also converts theses values to integrers and then update the dataframe.

```python
for i in range(len(df['Adults'])):
    if '€' in str(df['Adults'].iloc[i]):
        df['Adults'].iloc[i] = int(str(df['Adults'].iloc[i]).replace('€', ''))
        
for i in range(len(df['Children'])):
    if '€' in str(df['Children'].iloc[i]):
        df['Children'].iloc[i] = int(str(df['Children'].iloc[i]).replace('€', ''))
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='my_app')
def get_coordinates(city):
    try:
        location = geolocator.geocode(city)
        return location.latitude, location.longitude
    except:
        return None, None
df[['Latitude', 'Longitude']] = df['City'].apply(lambda x: pd.Series(get_coordinates(x)))
df
```
