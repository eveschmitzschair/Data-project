Loop pour avoir ville, attraction, prix brut et temps de visite

```python
path = "C:Users/flori/Desktop/Master 1/Q2/Data Management"
driver = webdriver.Chrome(executable_path=os.path.join(path, 'chromedriver'))
attractions_infos = []
url = "https://www.365.be/fr/recherche/?category=1062&range=150"
driver.get(url)
time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')

events = soup.find('div', class_='main-content__body').find('div', class_='row cards').find_all('article', class_='col-12 col-md-6 col-lg-6 col-xl-4')

for event in events:
    full_event_url = f"https://www.365.be{event.find('a')['href']}"
    driver.get(full_event_url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    try:
        attraction_name = soup.find('h1', class_="attraction-name").text.strip()
    except:
        attraction_name = "Not available"
    try:
        city_name = soup.find('span', {'itemprop': 'addressLocality'}).text.strip()
    except:
        city_name = "Not available"
    try:
        price_label = soup.find("span", class_="label", text="Prix").find_next_sibling("div").text.strip()
    except:
        price_label = "Not available"
    try:
        visiting_time_label = soup.find("span", class_="label", text="Durée de la visite").find_next_sibling("div").text.strip()
    except:
        visiting_time_label = "Not Available"
    attractions_infos.append({'Name' : attraction_name, 'City' : city_name, 'Price' : price_label, 'Visiting time' : visiting_time_label})
print(attractions_infos)
driver.quit()
```

```python
df = pd.DataFrame(attractions_infos)
df = df.sort_values(by=['City'])
df
```

```python
len(df)
```

```python
df["Price"] = df["Price"].astype(str)

# Define the function to extract price from the label
def extract_price(category, price_label):
    if category == "adultes":
        pattern = f"(adulte|adultes|adult).+?(\d+(?:[\.,]\d{{1,2}})?)\s*(?:€|\(.*?€\))"
    elif category == "enfants":
        pattern = f"(enfant|enfants).+?(\d+(?:[\.,]\d{{1,2}})?)\s*(?:€|\(.*?€\))"
    else:
        pattern = None
        
    if pattern:
        match = re.search(pattern, price_label, flags=re.IGNORECASE)
        if match:
            return match.group(2).replace(',', '.')
    
    
    if "gratuit" in price_label.lower() or "entrée gratuite" in price_label.lower() or "entrée libre" in price_label.lower():
        return 0
    if len(price_label.split()) == 1:
        return price_label.replace(',','.')    
    
    return None

# Extract prices for different categories
df["Adults"] = df["Price"].apply(lambda x: extract_price("adultes", x))
df["Children"] = df["Price"].apply(lambda x: extract_price("enfants", x))
```

```python
df.drop('Price', axis=1, inplace=True)
df
```

```python
for i in range(len(df['Adults'])):
    if '€' in str(df['Adults'][i]):
        df['Adults'][i] = int(str(df['Adults'][i]).replace('€', ''))
for i in range(len(df['Children'])):
    if '€' in str(df['Children'][i]):
        df['Children'][i] = int(str(df['Children'][i]).replace('€', ''))
df
```

```python
from geopy.geocoders import Nominatim

# Initialize the geolocator
geolocator = Nominatim(user_agent='my_app')

# Define a function to get the coordinates for a city
def get_coordinates(city):
    try:
        location = geolocator.geocode(city)
        return location.latitude, location.longitude
    except:
        return None, None

# Apply the function to the 'City' column and create two new columns for latitude and longitude
df[['Latitude', 'Longitude']] = df['City'].apply(lambda x: pd.Series(get_coordinates(x)))

# Print the updated dataframe
df
```
